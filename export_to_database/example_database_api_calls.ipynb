```python
# q = request.query_params.get("q", "").strip()
# page = max(int(request.query_params.get("page", 1)), 1)
# page_size = min(max(int(request.query_params.get("page_size", 20)), 1), 100)
# kit_id = request.query_params.get("lingoKitId")
# "pos": request.query_params.getlist("pos"),
# "topics": request.query_params.getlist("topics"),
# "gender": request.query_params.getlist("gender"),
# "plurality": request.query_params.getlist("plurality"),
# "article_type": request.query_params.getlist("article_type"),
# "preposition": request.query_params.getlist("preposition"),
# "lingo_kits": request.query_params.getlist("lingo_kits"),

def term_translations_search(
    q=None,
    include_entire_lemma_group=None,
    page=None,
    page_size=None,
    pos=None,
    topics=None,
    gender=None,
    plurality=None,
    article_type=None,
    preposition=None,
    lingo_kits=None,
    kit_id=None,
    auth=('stevie', 'lingokit2025!'),
):
    url = 'http://127.0.0.1:8000/api/search/term-translations/'
    query = '?'
    if q is not None:
        query += f'q={q}&'
    if include_entire_lemma_group is not None:
        assert(isinstance(include_entire_lemma_group, bool))
        query += f'include_entire_lemma_group={str(include_entire_lemma_group).lower()}&'
    if page is not None:
        assert(isinstance(page, int))
        query += f'page={page}&'
    if page_size is not None:
        assert(isinstance(page_size, int))
        query += f'page_size={page_size}&'
    if pos is not None:
        assert(isinstance(pos, list))
        for p in pos:
            assert(isinstance(p, str))
            query += f'pos={p}&'
    if topics is not None:
        assert(isinstance(topics, list))
        for t in topics:
            assert(isinstance(t, str))
            query += f'topics={t}&'
    if gender is not None:
        assert(isinstance(gender, list))
        for g in gender:
            assert(isinstance(g, str))
            query += f'gender={g}&'
    if plurality is not None:
        assert(isinstance(plurality, list))
        for p in plurality:
            assert(isinstance(p, str))
            query += f'plurality={p}&'
    if article_type is not None:
        assert(isinstance(article_type, list))
        for a in article_type:
            assert(isinstance(a, str))
            query += f'article_type={a}&'
    if preposition is not None:
        assert(isinstance(preposition, list))
        for p in preposition:
            assert(isinstance(p, str))
            query += f'preposition={p}&'
    if lingo_kits is not None:
        assert(isinstance(lingo_kits, list))
        for k in lingo_kits:
            assert(isinstance(k, int))
            query += f'lingo_kits={k}&'
    if kit_id is not None:
        assert(isinstance(kit_id, int))
        query += f'lingoKitId={kit_id}&'
    if query.endswith('&') or query.endswith('?'):
        query = query[:-1]
    response = requests.get(
        url + query,
        auth=auth,
    )
    return response


def term_translations_search_all_pages(
    q=None,
    include_entire_lemma_group=None,
    page_size=None,
    pos=None,
    topics=None,
    gender=None,
    plurality=None,
    article_type=None,
    preposition=None,
    lingo_kits=None,
    kit_id=None,
    auth=('stevie', 'lingokit2025!'),
):
    response_list = []
    page = 1
    while True:
        response = term_translations_search(
            q=q,
            include_entire_lemma_group=include_entire_lemma_group,
            page=page,
            page_size=page_size,
            pos=pos,
            topics=topics,
            gender=gender,
            plurality=plurality,
            article_type=article_type,
            preposition=preposition,
            lingo_kits=lingo_kits,
            kit_id=kit_id,
            auth=auth,
        )
        response = response.json()
        response_list.append(response)
        if response['hasMore'] is False:
            break
        page += 1
    return response_list
```

```python
from collections import defaultdict
import csv
from io import StringIO

BASE_URL = "http://127.0.0.1:8000"
AUTH = ("stevie", "lingokit2025!")

csv_data = """
term_italian,base_lemma_italian,base_lemma_english,part_of_speech,translation_english,topics,subtype,person_number,gender,plurality,mood,tense,is_comparative,is_superlative,is_compound,article_type,article_italian,added_particle_italian,preposition,notes,example_sentence_italian,example_sentence_english,english_audio_hash,english_duration_ms,italian_audio_hash,italian_duration_ms,pronunciation
abbastanza,abbastanza,enough,adverb,enough/quite/rather,"Numbers & Quantities,Pronouns & Determiners,Spotify Lessons",,,,,,,False,False,False,,,,,degree adverb,È abbastanza facile.,It is quite easy.,ec22a944133b60e6898488067584bf484d9b0135d2c3426f18bbf134b43cf431,3024.0,a26df5219355cfc1d190135def517689b74c407ecdf400c7eb6ffaa1fa85caad,1752.0,ahb-bah-STAN-tsah
altro,altro,other,adj,other (masculine),"Qualities & Descriptions,Concepts & Ideas,Pronouns & Determiners,Spotify Lessons",,,masculine,singular,,,False,False,False,,,,,,Prendo un altro caffè.,I am having another coffee.,e706a75b7ecf926cde61106f9f0d140671f032eebdbe446e22e0171024301d03,db853e333eb8ddf34c3176a930b6d2e1236e44631f63cfc5506b33849fb71e00,1968.0,1248.0,AHL-troh
altra,altro,other,adj,other (feminine),"Qualities & Descriptions,Concepts & Ideas,Pronouns & Determiners,Spotify Lessons",,,feminine,singular,,,False,False,False,,,,,,Porti altra acqua per favore.,Bring some more water please.,8b46fce3d70ecd48f1cb98e7726ac8fcd40d57cf19814541c92ecf0217246b67,9c498fd7dd492a85f76e9bc31f7af6e30dbef7be49fe60dea12aeb9a0f7ee50c,1776.0,1224.0,AHL-trah
altri,altro,other,adj,other (masculine plural),"Qualities & Descriptions,Concepts & Ideas,Pronouns & Determiners,Spotify Lessons",,,masculine,plural,,,False,False,False,,,,,,Ci sono altri problemi.,There are other problems.,6f5dc6909984121d78f37ed9a1b478d73daac15a719dcd87b47b42025b6ad7f5,cdc19d5925a143693abaa8fb8eac6de1ceab8467e0e5b17a9d2eb9c790bca2df,2424.0,1224.0,AHL-tree
altre,altro,other,adj,other (feminine plural),"Qualities & Descriptions,Concepts & Ideas,Pronouns & Determiners,Spotify Lessons",,,feminine,plural,,,False,False,False,,,,,,Ho letto altre storie.,I have read other stories.,42fc3932266bfbbfb84062f662749ec2db805f0178ff87e58979a56a8566b20b,44e21faec008513f40db45481630dd32cec191c27f6a44e0e118e9b6340495ff,2256.0,1200.0,AHL-treh
il,il,the,art,the (masculine singular),"Pronouns & Determiners,Spotify Lessons",definite,,masculine,singular,,,False,False,False,,,,,before most consonants (masculine singular),Prendo il treno.,I take the train.,d55e6647a4945740fd0fd28669a01e61634221c550d6571a1e0b8333daf6b6b9,3122120e20a5bada2cabaa9f9c90e0908a2e478038941d47b81da2edfef97d17,2496.0,1056.0,EEL
la,il,the,art,the (feminine singular),"Pronouns & Determiners,Spotify Lessons",definite,,feminine,singular,,,False,False,False,,,,,feminine singular before consonant,Pago la cena.,I pay for the dinner.,df6d10cb97d8003f70690a09236209674b9990974ae0450bd05cecf63c23aff8,271ad1c1578faac9fff37172607e627500ef3827657e6ae33e7fd031271ab43e,2328.0,960.0,LAH
i,il,the,art,the (masculine plural),"Pronouns & Determiners,Spotify Lessons",definite,,masculine,plural,,,False,False,False,,,,,masculine plural before consonant,Conosco i ragazzi.,I know the boys.,9170401a2d3c08a42d070f161a883e0650ad02a133452d58553fd847e4aa06f0,1e0dbb5aee81f17ad2d2140dcfb612ebe38ff243f98393752ed919c522886de9,2328.0,864.0,EE
le,il,the,art,the (feminine plural),"Pronouns & Determiners,Spotify Lessons",definite,,feminine,plural,,,False,False,False,,,,,feminine plural,Vedo le case.,I see the houses.,afc09f751cbedc5fc4158ff6da00a15e0673d1b8f8a04fcb00ef33924d031855,d5c4b7e2f41718617b6c7ab53b570e0627f2ed124b3e074359432bf063d7a80a,2160.0,960.0,LEH
lo,il,the,art,"the (masculine singular, before s+consonant, z, gn, ps, x, y)","Pronouns & Determiners,Spotify Lessons",definite,,masculine,singular,,,False,False,False,,,,,"before s+consonant, z, gn, ps, x, y (masculine singular)",Conosco lo studente.,I know the student.,440eef3fb9086b6f1f700b827b8e39e8e288218ff456c1c7d422b4f166705e8d,eca3948d282af162a6247d4e4d0452df14929f1c210794588671855b544bc9bd,7800.0,984.0,LOH
gli,il,the,art,"the (masculine plural, before vowels, s+consonants, z, gn, ps, x, y)","Pronouns & Determiners,Spotify Lessons",definite,,masculine,plural,,,False,False,False,,,,,"before vowels, s+consonant, z, gn, ps, x, y (masculine plural)",Vedo gli studenti.,I see the students.,bfe25a52891973085affa8b26433e1950110b66c5ff5385ce61ef5a39522d0bb,75cca62055819d90ece6bfe8607ad5191ba48c46ba8cecc9fd11b5f3e021c9a5,8640.0,1080.0,LYEE
l',il,the,art,"the (singular, before vowel)","Pronouns & Determiners,Spotify Lessons",definite,,,singular,,,False,False,False,,,,,before vowel (singular),Aspetto l'autobus.,I am waiting for the bus.,25896cdec1d5856dc7e29320d1e5ce041b61a2964358b60673bc8bc948ac9fbf,ef0035eea209c7a18822a203b843b28aa8f80a115fa60e9f9ede9f9eb52e3eee,2904.0,912.0,EL
"""

POS_MAP = {
    "adverb": "adv",
    "adv": "adv",
    "adj": "adj",
    "art": "art",
}
GENDER_MAP = {
    "masculine": "m",
    "feminine": "f",
    "n/a": "n/a",
}
PLURALITY_MAP = {
    "singular": "s",
    "plural": "p",
    "n/a": "n/a",
}

topic_cache = {}
baselemma_cache = {}
term_cache = {}
baselemma_topic_cache = set()
position_tracker = defaultdict(int)
expected_order = defaultdict(list)


def _extract_results(payload):
    if isinstance(payload, dict):
        results = payload.get('results')
        if isinstance(results, list):
            return results
        return []
    if isinstance(payload, list):
        return payload
    return []


def ensure_topic(name: str):
    name = name.strip()
    if not name:
        return None
    if name in topic_cache:
        return topic_cache[name]
    resp = requests.get(f"{BASE_URL}/api/topics/", params={"topic": name}, auth=AUTH)
    resp.raise_for_status()
    payload = resp.json()
    results = _extract_results(payload)
    if results:
        topic_cache[name] = results[0]["id"]
        return topic_cache[name]
    resp = requests.post(
        f"{BASE_URL}/api/topics/",
        json={"topic": name},
        auth=AUTH,
    )
    resp.raise_for_status()
    topic_cache[name] = resp.json()["id"]
    return topic_cache[name]


def ensure_baselemma(it: str, en: str):
    key = (it, en)
    if key in baselemma_cache:
        return baselemma_cache[key]
    resp = requests.get(
        f"{BASE_URL}/api/baselemmas/",
        params={"base_lemma_italian": it, "base_lemma_english": en},
        auth=AUTH,
    )
    resp.raise_for_status()
    payload = resp.json()
    # Paginated response returns results list; empty list means create new
    results = _extract_results(payload)
    if results:
        baselemma_cache[key] = results[0]["id"]
        return baselemma_cache[key]
    resp = requests.post(
        f"{BASE_URL}/api/baselemmas/",
        json={"base_lemma_italian": it, "base_lemma_english": en},
        auth=AUTH,
    )
    resp.raise_for_status()
    baselemma_cache[key] = resp.json()["id"]
    return baselemma_cache[key]


def ensure_term(term_it: str, audio_hash: str, pronunciation: str):
    key = term_it
    if key in term_cache:
        return term_cache[key]
    resp = requests.get(
        f"{BASE_URL}/api/terms/",
        params={"term_italian": term_it},
        auth=AUTH,
    )
    resp.raise_for_status()
    payload = resp.json()
    results = _extract_results(payload)
    if results:
        term_cache[key] = results[0]["id"]
        return term_cache[key]
    body = {"term_italian": term_it}
    if audio_hash:
        body["audio_hash_italian"] = audio_hash
    if pronunciation:
        body["pronunciation"] = pronunciation
    resp = requests.post(f"{BASE_URL}/api/terms/", json=body, auth=AUTH)
    resp.raise_for_status()
    term_cache[key] = resp.json()["id"]
    return term_cache[key]


def ensure_baselemma_topic(base_lemma_id: int, topic_id: int):
    if topic_id is None:
        return
    key = (base_lemma_id, topic_id)
    if key in baselemma_topic_cache:
        return
    resp = requests.post(
        f"{BASE_URL}/api/baselemma-topics/",
        json={"base_lemma": base_lemma_id, "topic": topic_id},
        auth=AUTH,
    )
    if resp.status_code not in (200, 201):
        resp.raise_for_status()
    baselemma_topic_cache.add(key)


def ensure_term_translation(payload: dict):
    params = {
        "term": payload["term"],
        "base_lemma": payload["base_lemma"],
        "translation_english": payload["translation_english"],
        "part_of_speech": payload["part_of_speech"],
    }
    resp = requests.get(
        f"{BASE_URL}/api/term-translations/",
        params=params,
        auth=AUTH,
    )
    resp.raise_for_status()
    existing = resp.json()
    existing_results = _extract_results(existing)
    if existing_results:
        return existing_results[0]["id"]
    resp = requests.post(
        f"{BASE_URL}/api/term-translations/",
        json=payload,
        auth=AUTH,
    )
    resp.raise_for_status()
    return resp.json()["id"]


reader = csv.DictReader(StringIO(csv_data.strip()))
created_term_translation_ids = []
for row in reader:
    base_it = row["base_lemma_italian"].strip()
    base_en = row["base_lemma_english"].strip()
    term_it = row["term_italian"].strip()
    translation_en = row["translation_english"].strip()
    key = (base_it, base_en)
    expected_order[key].append((term_it, translation_en))

    base_lemma_id = ensure_baselemma(base_it, base_en)
    term_id = ensure_term(
        term_it,
        row["italian_audio_hash"].strip(),
        row["pronunciation"].strip(),
    )

    topics = [t.strip() for t in row["topics"].split(",") if t.strip()]
    for topic in topics:
        topic_id = ensure_topic(topic)
        ensure_baselemma_topic(base_lemma_id, topic_id)

    pos_key = row["part_of_speech"].strip().lower()
    pos_value = POS_MAP.get(pos_key)
    if not pos_value:
        raise ValueError(f"Unhandled part_of_speech: {row['part_of_speech']}")

    payload = {
        "term": term_id,
        "base_lemma": base_lemma_id,
        "translation_english": translation_en,
        "part_of_speech": pos_value,
        "position": position_tracker[key],
    }
    position_tracker[key] += 1

    gender_value = row["gender"].strip().lower()
    gender = GENDER_MAP.get(gender_value)
    if gender:
        payload["gender"] = gender

    plurality_value = row["plurality"].strip().lower()
    plurality = PLURALITY_MAP.get(plurality_value)
    if plurality:
        payload["plurality"] = plurality

    if row["article_type"].strip():
        payload["article_type"] = row["article_type"].strip()
    if row["article_italian"].strip():
        payload["article_italian"] = row["article_italian"].strip()
    if row["preposition"].strip():
        payload["preposition"] = row["preposition"].strip()
    if row["example_sentence_english"].strip():
        payload["example_sentence_english"] = row["example_sentence_english"].strip()
    if row["example_sentence_italian"].strip():
        payload["example_sentence_italian"] = row["example_sentence_italian"].strip()
    if row["english_audio_hash"].strip():
        payload["audio_hash_english"] = row["english_audio_hash"].strip()
    if row["notes"].strip():
        payload["notes"] = row["notes"].strip()

    tt_id = ensure_term_translation(payload)
    created_term_translation_ids.append(tt_id)

pages = term_translations_search_all_pages(page_size=50)
observed_keys = set()
for page in pages:
    for group in page["groups"]:
        key = (group["baseLemma"]["it"], group["baseLemma"]["en"])
        observed_keys.add(key)
        if key in expected_order:
            observed = [
                (item["term"]["it"], item["translation_en"]) for item in group["items"]
            ]
            assert observed == expected_order[key], (
                f"Unexpected order for {key}: {observed} != {expected_order[key]}"
            )

missing_keys = set(expected_order) - observed_keys
if missing_keys:
    details = ", ".join(f"{it} — {en}" for it, en in missing_keys)
    raise AssertionError(f"Missing lemma groups in API response: {details}")

print(
    f"Created/verified {len(created_term_translation_ids)} term translations "
    f"across {len(expected_order)} base lemma groups."
)
```
